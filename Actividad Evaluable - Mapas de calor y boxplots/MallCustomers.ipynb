```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Carga los datos usando tu lector de csv o con pandas. Es recomendable hacerlo con pandas.
df = pd.read_csv(r"C:\Users\52331\Desktop\Tecnologico de Monterrey\3er Semestre\ArteAnalitica\Actividad Evaluable - Mapas de calor y boxplots\Mall_Customers.csv")
df.head()
```

```python
sns.boxplot(data=df, x="Gender", y="Age")
plt.title("Distribución de Edad por Género")
plt.ylabel("Edad")
plt.xlabel("Género")
plt.show()

sns.boxplot(data=df, x="Gender", y="Annual Income (k$)")
plt.title("Ingresos Anuales (k$) por Género")
plt.ylabel("Annual Income (k$)")
plt.xlabel("Género")
plt.show()

sns.boxplot(data=df, x="Gender", y="Spending Score (1-100)")
plt.title("Spending Score por Género")
plt.ylabel("Spending Score (1-100)")
plt.xlabel("Género")
plt.show()

sns.boxplot(data=df, x='income_bin', y='Age')
plt.title('Edad por grupo de Ingreso')
plt.xlabel('Income bin')
plt.ylabel('Edad')
plt.show()

sns.boxplot(data=df, x='income_bin', y="Spending Score (1-100)")
plt.title('Spending Score por grupo de Ingreso (quantiles)')
plt.xlabel('Income binn')
plt.ylabel('Spending Score (1-100)')
plt.show()

sns.boxplot(data=df, x='spend_bin', y='income_bin')
plt.title('Annual Income por grupo de Spending Score (quantiles)')
plt.xlabel('Spending Score')
plt.ylabel('Annual Income (k$)')
plt.show()
```

```python
sns.histplot(df["Age"], kde=True, color="skyblue")
plt.title("Histograma de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

sns.histplot(df["Annual Income (k$)"], kde=True, color="lightgreen")
plt.title("Histograma de Ingresos Anuales (k$)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Frecuencia")
plt.show()

sns.histplot(df["Spending Score (1-100)"], kde=True, color="salmon")
plt.title("Histograma del Spending Score (1-100)")
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df["Gender_num"] = df["Gender"].map({"Male": 0, "Female": 1})

columnas = ["Gender_num", "Age", "Annual Income (k$)", "Spending Score (1-100)"]
corr = df[columnas].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de calor")
plt.show()
```
